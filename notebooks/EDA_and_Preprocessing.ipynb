```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, LabelEncoder, StandardScaler

# Dataset'i yükle
df = pd.read_excel("../data/dataset.xlsx")  # kendi dosya adını yaz
df.head()
```

```python
print("Veri boyutu:", df.shape)
df.info()
df.describe()
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(df.isnull(), cbar=False, cmap="viridis")
plt.title("Eksik Veriler")
plt.show()

print(df.isnull().sum())
```

```python
# TedaviSuresi dağılımı
sns.histplot(df["TedaviSuresi"], bins=30, kde=True)
plt.title("Tedavi Süresi Dağılımı")
plt.show()

# Kategorik sütunlar
categorical_cols = ["Cinsiyet","KanGrubu","Uyruk","Bolum","TedaviAdi"]
for col in categorical_cols:
    plt.figure(figsize=(8,4))
    sns.countplot(data=df, x=col)
    plt.title(f"{col} Dağılımı")
    plt.xticks(rotation=45)
    plt.show()
```

```python
# 'Dakika' yazısını kaldır ve sayıya çevir
df['UygulamaSuresi'] = df['UygulamaSuresi'].str.replace('Dakika','', regex=False).str.strip()
df['UygulamaSuresi'] = pd.to_numeric(df['UygulamaSuresi'], errors='coerce')

# Yas sütununu sayıya çevir (gerekirse)
df['Yas'] = pd.to_numeric(df['Yas'], errors='coerce')
```

```python
from sklearn.impute import SimpleImputer

# Sayısal sütunlar
num_cols = ["Yas", "UygulamaSuresi"]
num_imputer = SimpleImputer(strategy="median")
df[num_cols] = num_imputer.fit_transform(df[num_cols])

# Kategorik sütunlar
cat_cols = ["Cinsiyet","KanGrubu","Uyruk","Bolum","Alerji","KronikHastalik","Tanilar","TedaviAdi","UygulamaYerleri"]
cat_imputer = SimpleImputer(strategy="constant", fill_value="Bilinmiyor")
df[cat_cols] = cat_imputer.fit_transform(df[cat_cols])

print("Eksik değerler dolduruldu.")
```

```python
# Label Encoding (Cinsiyet gibi 2 sınıf)
le = LabelEncoder()
df["Cinsiyet"] = le.fit_transform(df["Cinsiyet"])

# OneHotEncoding (Bolum, TedaviAdi, KanGrubu vb.)
ohe_cols = ["KanGrubu","Uyruk","Bolum","Alerji","KronikHastalik","Tanilar","TedaviAdi","UygulamaYerleri"]
df = pd.get_dummies(df, columns=ohe_cols, drop_first=True)

df.head()
```

```python
scaler = StandardScaler()
df[["Yas","UygulamaSuresi"]] = scaler.fit_transform(df[["Yas","UygulamaSuresi"]])

df.head()
```

```python
# Modellemeye hazır CSV
df.to_csv("../data/dataset_model_ready.csv", index=False)
print("Modellemeye hazır dataset kaydedildi: dataset_model_ready.csv")
```
